```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.DiGraph()

#Definiranje spojnica
G.add_edge(0,1)
G.add_edge(0,4)
G.add_edge(0,5)

G.add_edge(2,0)
G.add_edge(2,5)
G.add_edge(2,7)

G.add_edge(3,0)
G.add_edge(3,4)

G.add_edge(4,7)

G.add_edge(5,3)
G.add_edge(5,4)
G.add_edge(5,9)

G.add_edge(6,1)
G.add_edge(6,2)
G.add_edge(6,7)

G.add_edge(7,1)

G.add_edge(8,1)
G.add_edge(8,2)
G.add_edge(8,4)
G.add_edge(8,9)

#Definiranje imena spojnica
G.edge[0][1]['Tezina']='4'
G.edge[0][4]['Tezina']='2'
G.edge[0][5]['Tezina']='2'

G.edge[2][0]['Tezina']='2'
G.edge[2][5]['Tezina']='1'
G.edge[2][5]['Tezina']='3'

G.edge[3][0]['Tezina']='2'
G.edge[3][4]['Tezina']='2'

G.edge[4][7]['Tezina']='3'

G.edge[5][3]['Tezina']='1'
G.edge[5][4]['Tezina']='2'
G.edge[5][9]['Tezina']='1'

G.edge[6][1]['Tezina']='1'
G.edge[6][2]['Tezina']='2'
G.edge[6][7]['Tezina']='2'

G.edge[7][1]['Tezina']='3'

G.edge[8][1]['Tezina']='1'
G.edge[8][2]['Tezina']='3'
G.edge[8][4]['Tezina']='3'
G.edge[8][9]['Tezina']='2'

#Definiranje imena kao atribut
G.node[0]['Ime']='LOK'
G.node[1]['Ime']='DIN'
G.node[2]['Ime']='NK ZG'
G.node[3]['Ime']='OSJ'
G.node[4]['Ime']='INT Z.'
G.node[5]['Ime']='HJD S'
G.node[6]['Ime']='RNK SPL'
G.node[7]['Ime']='SL B.'
G.node[8]['Ime']='IST 61'
G.node[9]['Ime']='RJ'


pozicija = nx.spring_layout(G)

nx.draw(G,pozicija)

#Crtanje imena
imena_label = nx.get_node_attributes(G,'Ime')
nx.draw_networkx_labels(G,pozicija,labels = imena_label)

#Crtanje imena spojnica
spojnica_label = nx.get_edge_attributes(G,'Tezina')
nx.draw_networkx_edge_labels(G,pozicija, spojnica_label)

pr=nx.pagerank(G, alpha=0.85)
print(pr)
plt.show()
```
