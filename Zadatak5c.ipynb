```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.DiGraph()

#Definiranje spojnica
G.add_edge(0,1)

G.add_edge(1,2)
G.add_edge(1,3)

G.add_edge(2,0)
G.add_edge(2,1)
G.add_edge(2,2)

#Definiranje imena spojnica
G.edge[0][1]['Tezina']='1'

G.edge[1][2]['Tezina']='1/2'
G.edge[1][3]['Tezina']='1/2'

G.edge[2][0]['Tezina']='1/3'
G.edge[2][1]['Tezina']='1/3'
G.edge[2][2]['Tezina']='1/3'

#Definiranje imena kao atribut
G.node[0]['Ime']='P1'
G.node[1]['Ime']='P2'
G.node[2]['Ime']='P3'
G.node[3]['Ime']='P4'

pozicija = nx.spring_layout(G)

nx.draw(G,pozicija)

#Crtanje imena
imena_label = nx.get_node_attributes(G,'Ime')
nx.draw_networkx_labels(G,pozicija,labels = imena_label)

#Crtanje imena spojnica
spojnica_label = nx.get_edge_attributes(G,'Tezina')
nx.draw_networkx_edge_labels(G,pozicija, spojnica_label)

pr=nx.pagerank(G, alpha=0.85)
print(pr)
plt.show()
```
